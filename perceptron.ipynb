```python
threshold = 1.7
inputs = [1, 0, 1, 0, 1]
weights = [0.7, 0.6, 0.5, 0.3, 0.4]
```

```python
def perceptron(): 
    sum = 0
    for i in range(len(inputs)):
      sum += inputs[i] * weights[i]

    activate = (sum > threshold)
    return activate
```

```python
if(perceptron()):
    print('1')
else:
    print(0)
```
