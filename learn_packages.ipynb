```python
import statistics as stats
import numpy as np
import math
```

```python
ages = [18, 19, 20, 21, 22, 22, 23, 24, 25, 25, 25]
```

```python
# Statistics package example

print("Mean:", stats.mean(ages))
print("Median:", stats.median(ages))
print("Mode:", stats.mode(ages))
print("Variance:", stats.variance(ages))
print("Standard Deviation:", stats.stdev(ages))
```

```python
x = 16
print("Square Root of 16:", math.sqrt(x))
print("Power: 2^3", math.pow(2, 3))
print("100 Log base 10:", math.log10(100))
print("Sine (90°):", math.sin(math.radians(90)))
print("Factorial of 5:", math.factorial(5))
```

```python
data = np.array([10, 20, 30, 40, 50])
print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Standard Deviation:", np.std(data))
print("Sum:", np.sum(data))


matrix = np.array([[1, 2, 3], [4, 5, 6]])
print("Matrix Shape:", matrix.shape)
print("Transpose:\n", matrix.T)
```
